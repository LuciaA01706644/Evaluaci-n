```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_csv('DF_Mexico.csv')
df  
```

```python
#Reemplazamos valores nulos del dataframe con "bfill"
df1 =df.fillna(method="bfill") 
df1
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=df1.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores nulos del dataframe con "mean"
#Realizamos una copia del dataframe
df2=df.copy()
df2=df2.fillna(round(df.mean(),1))
df2
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
df2.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=df2
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

Se utilizó este método para eliminar valores atípicos debido a que es menos senisble y ayuda a mantener la forma general de los datos.


```python
#Encontramos Ouliers del Dataframe
outliers= df2[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
df3= df2[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df3
```

```python
#Eliminamos columnas innecesarias
df4= df3.drop(['listing_url','last_scraped','source'], axis=1)
df4
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=df4.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=df4.copy()
data_clean=df.fillna(round(df.mean(),1))
data_clean
```

Se utilizó este método con el propósito de que los datos faltantes, sean reemplazados por datos "reales" o que si sea probable que estén y no muestre más valores atípicos.

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```
